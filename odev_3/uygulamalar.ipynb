```python
class banka(): #yeni bir banka sinifi olusturup ortak banka ozelligi ekliyorum
    hangi_banka = "yapikredi"

    def __init__(self, isim, soyisim, hesap_no, sifre, bakiye): #ozelliklerimi tanimliyorum
        self.isim = isim
        self.soyisim = soyisim
        self.hesap_no = hesap_no
        self.sifre = sifre
        self.bakiye = bakiye

    def bakiye_goruntule(self): #bakiye goruntuleme methodunu tanimliyorum
        numara = input("Hesap numaranizi giriniz: ") #hesap no istiyorum
        if numara != self.hesap_no:
            print("Yanlis veya gecersiz bir numara girdiniz.") #yanlis girildiginde guvenlik onlemi nedeniyle uygulama duruyor.
            return #durdugu yer

        print("Hesap numaraniz dogrudur.") #dogruysa devam ediyor

        pw = input("Sifrenizi giriniz: ") #sifre istiyorum
        if pw != self.sifre:
            print("Yanlis veya gecersiz bir sifre girdiniz.")  #yanlissa uygulama duruyor
            return

        print("Sifreniz dogrudur.")
        print("Guncel bakiyeniz: {}".format(self.bakiye)) #kullanici bakiyesini gosteriyorum.

    def para_cekme(self):
        numara = input("Hesap numaranizi giriniz: ")
        if numara != self.hesap_no:
            print("Yanlis veya gecersiz bir numara girdiniz.")
            return

        print("Hesap numaraniz dogrudur.")

        pw = input("Sifrenizi giriniz: ")
        if pw != self.sifre:
            print("Yanlis veya gecersiz bir sifre girdiniz.")
            return

        print("Sifreniz dogrudur.")

        cekim_miktari = input("Ne kadar para cekmek istiyorsunuz?")
        if float(cekim_miktari) > self.bakiye:
            print("Hesabinizdaki bakiye yetersiz.") #eger cekim miktari bakiyeden fazlaysa uyari mesaji gelip uygulama duruyor
            return

        self.bakiye = self.bakiye - float(cekim_miktari) #guncel miktari tanimliyorum, inputu str aldigimiz icin float olarak degerlendirilmesini sagliyorum
        print("Islem basariyla gerceklesti. Guncel bakiye: {}".format(self.bakiye)) #cıktı olarak gosteriyorum

    def para_yatirma(self):
        numara = input("Hesap numaranizi giriniz: ")
        if numara != self.hesap_no:
            print("Yanlis veya gecersiz bir numara girdiniz.")
            return

        print("Hesap numaraniz dogrudur.")

        pw = input("Sifrenizi giriniz: ")
        if pw != self.sifre:
            print("Yanlis veya gecersiz bir sifre girdiniz.")
            return

        print("Sifreniz dogrudur.")

        yatirma_miktari = input("Yatirmak istediginiz miktari giriniz:")
        self.bakiye = self.bakiye + float(yatirma_miktari)
        print("Islem basariyla gerceklesti. Guncel bakiyeniz: {}".format(self.bakiye)) #ayni mantikla para yatirma uygulamasi.

        

kullanici = banka("Isik", "Tuysuz", "24016020", "2402", 5000)
```
